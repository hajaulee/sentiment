```python
import csv
import numpy as np
import emoji
import pandas as pd
import matplotlib.pyplot as plt
def read_csv(filename = 'emojify_data.csv'):
    phrase = []
    emoji = []

    with open (filename) as csvDataFile:
        csvReader = csv.reader(csvDataFile)

        for row in csvReader:
            phrase.append(row[0])
            emoji.append(row[1])

    X = np.asarray(phrase)
    Y = np.asarray(emoji, dtype=int)

    return X, Y
```

```python
import re
def read_crash_txt(filename='train.crash.txt'):
#     seg = Segmenter('./data')
    phrase = []
    emoji = []
    labels_list = ['0', '1'] 
    cmt = ""
    with open (filename) as file:
        for line in file:
            if line[0] == '"':
                cmt = line
            elif line[0] in labels_list:
                cmt = ' '.join(re.findall(r'[\w]+', cmt)).lower()
#                 cmt = seg.graph_dynamic_programing(cmt)
                phrase.append(cmt)
                emoji.append(int(line[0]))
                cmt = ""
            elif cmt != "":
                cmt += line
            

    X = np.asarray(phrase)
    Y = np.asarray(emoji, dtype=int)

    return X, Y
```

```python
X, Y = read_csv('imdb_labelled.txt.csv')
```

```python
X, Y = read_crash_txt()
```

```python
from sklearn.model_selection import train_test_split    
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from collections import Counter
class Classifier:
    def __init__(self):
        pass
    def fit(self, _X, _Y):
        def text_counter(sentences_list):
            all_text2 = ' '.join(sentences_list).lower()
            words = all_text2.split()
            count_words = Counter(words)
            return count_words
        _po, _ne = [], []

        for i in range(len(_X)):
            if Y[i] == 0:
                _ne.append(_X[i])
            else:
                _po.append(_X[i])
        f_po = text_counter(_po)
        f_ne = text_counter(_ne)
        self.only_ne = list(filter(lambda x: x not in f_po.keys(), f_ne.keys()))
        self.only_po = list(filter(lambda x: x not in f_ne.keys(), f_po.keys()))

    def predict(self, _X):
        def predict(x, ne_dict, po_dict):
            words_list = x.lower().split(' ')
            words_list.reverse()
            for word in x.lower().split(' '):
                if word in ne_dict: return 0
                if word in po_dict: return 1
            return 0
        Y_pred = [predict(x, self.only_ne, self.only_po) for x in _X]
        return Y_pred
    
    def evaluate(self, _X, _Y):
        Y_pred = self.predict(_X)
        acc = 0
        for i in range(len(Y_pred)):
            if Y_pred[i] == _Y[i]:acc+=1
        print("Accuracy: {}%".format(100*acc/len(Y_pred)))
        
```
